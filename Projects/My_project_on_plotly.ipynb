## My Project On Plotly


```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
```

```python
df = pd.read_csv("C:\\Users\\DELL\\OneDrive\\Desktop\\Bhargav's visual python\superstore_data.csv")
df
```

# MY 1ST GRAPH
Here, I will present what are the marital statuses of people on the basis of birth year

```python
figure1 = px.scatter(df,x='Year_Birth',y='Marital_Status',title="Year_Birth vs Marital_Status")
figure1.show()
```

## MY 2nd Graph
Here, I will present in which year most people are still single.

```python
df2 = df[df["Marital_Status"] == "Single"]
df2.groupby("Year_Birth")
figure2 = px.bar(df2,x=df2["Year_Birth"].drop_duplicates(),y=df2["Year_Birth"].value_counts(),labels={"x":"Birth Year","y":"Count of People born in that year"},title="Year_Birth vs Marital_status")
figure2.show()
```

# MY 3rd Graph
Here,I will present count of people by birth year who done phd in education are divorced and have income more than 50k.

```python
df3 = df[df["Marital_Status"]=="Divorced"][df["Education"]=="PhD"][df["Income"]>50000]
print(df3)

figure3 = px.bar(df3,x=df["Year_Birth"].drop_duplicates(),y=df["Year_Birth"].value_counts(),labels={"x":"Birth year","y":"Count of people born in that year"},title="DIVORCED PhD People by Birth year")
figure3.show()
```

# MY 4th Graph
Here,I will present the visualization of given datasets by three colomns.1.birth year,2.education,3.income

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

figure5 = make_subplots(rows=1,cols=2)

figure5.add_trace(go.scatter(df,x="Year_Birth",y="Education",row=1,col=1))

figure5.add_trace(go.scatter(df,x="Year_Birth",y="Marital_Status",row=1,col=2))

figure5.show()
```
